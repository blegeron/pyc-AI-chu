```python
import pandas as pd

# Load the CSV file into a DataFrame
csv_path = "/home/user_jodert/code/joderthibaud/Personal/pyc-AI-chuvTHJ/data/poke_battle_details.csv"
df = pd.read_csv(csv_path)
df
```

```python

# Group by 'poke_battle_ref' and calculate the average win rate
pokemon_win_rates = df.groupby('poke_battle_ref').agg(
    avg_win_rate=('winner', 'mean'),
    num_battles=('winner', 'count')
).reset_index().sort_values(by='avg_win_rate', ascending=False)
```
